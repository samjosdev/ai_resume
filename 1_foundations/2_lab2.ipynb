````python
request = "Please come up with a challenging, nuanced question that I can ask a number of LLMs to evaluate their intelligence. "
request += "Answer only with the question, no explanation."
messages = [{"role": "user", "content": request}]
````

````python
messages
````

````python
openai = OpenAI()
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages,
)
question = response.choices[0].message.content
print(question)
````

````python
competitors = []
answers = []
messages = [{"role": "user", "content": question}]
````

````python
# The API we know well

model_name = "gpt-4o-mini"

response = openai.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
````

````python
# Anthropic has a slightly different API, and Max Tokens is required

claude = Anthropic()
response = claude.messages.create(model=cost_effective_claude_model, messages=messages, max_tokens=1000)
answer = response.content[0].text

display(Markdown(answer))
competitors.append(cost_effective_claude_model)
answers.append(answer)
````

````python
gemini = OpenAI(api_key=google_api_key, base_url="https://generativelanguage.googleapis.com/v1beta/openai/")
model_name = "gemini-2.5-flash-preview-05-20"

response = gemini.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
````

````python
deepseek = OpenAI(api_key=deepseek_api_key, base_url="https://api.deepseek.com/v1")
model_name = "deepseek-chat"

response = deepseek.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
````

````python
groq = OpenAI(api_key=groq_api_key, base_url="https://api.groq.com/openai/v1")
model_name = "llama-3.3-70b-versatile"

response = groq.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
````

````python
# So where are we?

print(competitors)
print(answers)
````

````python
# It's nice to know how to use "zip"
for competitor, answer in zip(competitors, answers):
    print(f"Competitor: {competitor}\n\n{answer}")
````

````python
# Let's bring this together - note the use of "enumerate"

together = ""
for index, answer in enumerate(answers):
    together += f"# Response from competitor {index+1}\n\n"
    together += answer + "\n\n"
````

````python
print(together)
````

````python
judge = f"""You are judging a competition between {len(competitors)} competitors.
Each model has been given this question:

{question}

Your job is to evaluate each response for clarity and strength of argument, and rank them in order of best to worst.
Respond with JSON, and only JSON, with the following format:
{{"results": ["best competitor number", "second best competitor number", "third best competitor number", ...]}}

Here are the responses from each competitor:

{together}

Now respond with the JSON with the ranked order of the competitors, nothing else. Do not include markdown formatting or code blocks."""
````

````python
print(judge)
````

````python
judge_messages = [{"role": "user", "content": judge}]
````

````python
# Judgement time!

openai = OpenAI()
response = openai.chat.completions.create(
    model="o3-mini",
    messages=judge_messages,
)
results = response.choices[0].message.content
print(results)
````

````python
print(competitors)
````

````python
# OK let's turn this into results!

results_dict = json.loads(results)
# print (results_dict)
ranks = results_dict["results"]
# print (ranks)
for index, result in enumerate(ranks):
    competitor = competitors[int(result)-1]
    print(f"Rank {index+1}: {competitor}")
````

# What pattern did this use?
Both had predefined workflow. So, they fall under the workflow design. Both can be classified as agents,as there were multiple LLM calls made.
Possible Agent frameworks:
- Workflow parallelization:
    - Same tasks were sent to multiple LLMs to compare
- Evaluator Optimizer
    - LLM as a judge was used here to evaluate the results.

````python
#lets prompt llm to give us criterias to judge the answers
criteria_prompt = f''' This is the input question: 
"new ethical framework for artificial intelligence that addresses the potential risks and benefits of autonomous decision-making in critical situations"
I need you to give me criterias to judge the responses by
Respond with JSON, and only JSON, with the following format:
{{"criteria": ["criteria 1", "criteria 2", ...]}}
Do not include markdown formatting or code blocks.
do not include ```json tag
'''
print(criteria_prompt)
````

````python
judging_criteria_message = [{
    "role":"user",
    "content":criteria_prompt
}]
````

````python
model_name = "gemini-2.5-flash-preview-05-20"
response = gemini.chat.completions.create(model=model_name, messages=judging_criteria_message)
answer = response.choices[0].message.content

criteria_dict = json.loads(answer)
criterias = criteria_dict["criteria"]
print (f'Criteria count: {len(criterias)}')
for index, value in enumerate(criterias):
    print (f'criteria # {index+1}: {value}')
````

````python
evaluate_based_on_criteria_prompt = f'''
You are judging a competition between {len(competitors)} competitors.
Each model has been given this question:

{question}

Your job is to evaluate each response for each of the following criteria:
{criterias}

Grade each competitor on a scale from 1 to 10 for each criterion (1 = lowest, 10 = perfect).

Respond in **pure JSON** (no code blocks, no markdown), with the following structure:

{{
  "criteria_grades": {{
    "criteria_1 (this is a tag. dont fill it with actual criteria_1)": {{
      "criteria": "<description of criteria 1>",
      "grades": {{
        "competitor_1": <grade>,
        "competitor_2": <grade>,
        ...
      }}
    }},
    "criteria_2 (this is a tag. dont fill it with actual criteria_2)": {{
      "criteria": "<description of criteria 2>",
      "grades": {{
        "competitor_1": <grade>,
        "competitor_2": <grade>,
        ...
      }}
    }},
    ...
  }}
}}

Here are the responses from each competitor:

{together}

Now respond with only the JSON output as described above. Do not include any extra text, explanations, or formatting.
'''
````

````python
print (evaluate_based_on_criteria_prompt)
````

````python
judge_message = [{
    "role":"user",
    "content":evaluate_based_on_criteria_prompt
}]
````

````python
response = openai.chat.completions.create(
    model="o3-mini",
    messages=judge_message,
)
open_ai_grades = response.choices[0].message.content
# print (open_ai_grades)
````

````python
response = claude.messages.create(model=cost_effective_claude_model, messages=judge_message, max_tokens=10000)
anthropic_grades = response.content[0].text
# print (anthropic_grades)
````

````python
model_name = "gemini-2.5-flash-preview-05-20"

response = gemini.chat.completions.create(model=model_name, messages=judge_message)
gemini_grades = response.choices[0].message.content
# print (gemini_grades)
````

````python
synthesis_prompt = f'''
You are given three grading JSONs from different judges (Gemini, Anthropic, OpenAI), each evaluating multiple competitors across several criteria.

Each JSON follows similar format.
{{
  "grades": {{
    "criteria_1": {{
      "criteria": "<description of criteria 1>",
      "grades": {{
        "competitor_1": <grade>,
        "competitor_2": <grade>
      }}
    }},
    "criteria_2": {{
      "criteria": "<description of criteria 2>",
      "grades": {{
        "competitor_1": <grade>,
        "competitor_2": <grade>
      }}
    }},
    ...
  }}
}}

### Your tasks:
These are the competitors:
{competitors}. competitor in index 0, is competitor 1, index 1 is competitor 2 etc.,
Here are the criterias:
{criterias}
output should show the model names and their ranking and not use generic competitor_1, 2 etc.,
1. Combine all three JSONs.
2. For each criterion:
   - Show the average grade for each competitor.
   - Highlight any large differences in judge scores.
3. Provide a brief summary of:
   - Which competitor performed best overall.
   - Who excelled in specific criteria.
   - Any notable disagreements or variations between the judges.
4. Display the results in a markdown table:
   - Rows = Each criterion + "Overall Average"
   - Columns = Each competitor
   - Cells = Average score (rounded to one decimal)

### Output Format:

- Summary (1–2 paragraphs)
- Markdown table:
Criteria	Competitor_1	Competitor_2
Clarity	8.7	7.3
Relevance	9.0	8.0
Overall Average	8.85	7.65
each row is for each criteria. Don't add any new criteria or breakdown criteria into multiple rows

Here are the JSONs:
*** Gemini ***
{gemini_grades}

**Anthropic:**
{anthropic_grades}

**OpenAI:**
{open_ai_grades}

Only output the summary and the markdown table.
number of rows in the table should match the number of criterias.
'''
````

````python
synthesis_message= [{
    "role":"user",
    "content":synthesis_prompt
}]
````

````python
model_name = "gemini-2.5-flash-preview-05-20"

response = gemini.chat.completions.create(model=model_name, messages=synthesis_message)
gemini_synthesis_response = response.choices[0].message.content
# display (Markdown(gemini_synthesis_response))
````

````python
response = claude.messages.create(model=cost_effective_claude_model, messages=synthesis_message, max_tokens=10000)
claude_synthesis_response = response.content[0].text
# display (Markdown(claude_synthesis_response))
````

````python
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=synthesis_message,
)
gpt_synthesis_response = response.choices[0].message.content
# display (Markdown(gpt_synthesis_response))
````

````python
final_response_prompt = f''' take these three inputs 
{gemini_synthesis_response}
{claude_synthesis_response}
{gpt_synthesis_response}

combine them into a common response with the summarized table in the end.
'''
final_message = [{
    "role":"user",
    "content": final_response_prompt
}]
````

````python
model_name = "gemini-2.5-flash-preview-05-20"

response = gemini.chat.completions.create(model=model_name, messages=final_message)
final_judgement = response.choices[0].message.content
display (Markdown(final_judgement))
````
